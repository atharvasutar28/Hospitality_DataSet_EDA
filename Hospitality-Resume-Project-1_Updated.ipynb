# Important Insights from the DataSets

```python
#Total Revenue Realized

total_revenue_realized = bookings_hotels_df['revenue_realized'].sum()
total_revenue_realized
```

##### Total revenue realized is way lower than revenue generated.

```python
#Total Bookings

total_bookings = bookings_hotels_df['booking_id'].nunique()
total_bookings
```

```python
#Total Capacity

total_capacity = aggregated_bookings_date_df['capacity'].sum()
total_capacity
```

```python
#Total Successful Bookings

total_successful_bookings = aggregated_bookings_date_df['successful_bookings'].sum()
total_successful_bookings
```

```python
#Occupancy %

occupancy_per = (total_successful_bookings / total_capacity) * 100
occupancy_per
```

##### Roughly 58% of rooms are booked and 42% of rooms are empty.

```python
#Average Rating

avg_rating = bookings_hotels_df['ratings_given'].sum() / total_bookings
avg_rating
```

##### Average rating given by a customer is roughly 1.5 on the scale of 5.
##### Service need to be look on.

```python
#Total Cancelled Bookings

total_cancelled_bookings = bookings_hotels_df.booking_status.str.count("Cancelled").sum()
total_cancelled_bookings
```

```python
#Cancellation %

cancellation_per = (total_cancelled_bookings / total_successful_bookings) * 100
cancellation_per
```

##### Cancellation of bookings are almost 25% against the total rooms booked. Something is wrong here, customers might be getting a better deals from our compititors.

```python
# Total Checked Out

total_checked_out = bookings_hotels_df.booking_status.str.count("Checked Out").sum()
total_checked_out
```

```python
#Total No Show Bookings

total_no_show_bookings = bookings_hotels_df.booking_status.str.count("No Show").sum()
total_no_show_bookings
```

```python
#No Show Bookings %

no_show_bookings_per = (total_no_show_bookings / total_successful_bookings) * 100
no_show_bookings_per
```

##### No Show Bookings in the Dataset is almost 5%. Nothing is wrong here, the number is small to be look on.

```python
#Booking Percentage of each Platform

for i in bookings_hotels_df['booking_platform'].unique():
    print(i, ":", (bookings_hotels_df.booking_platform.str.count(i).sum() / total_successful_bookings) * 100)
```

##### MakeYourTrip and LogTrip are engaging more customers than any others, others are also attracting a large chunk of customers.

```python
#Booking Percentage of each Room Class

for i in bookings_hotels_df['room_class'].unique():
    print(i, ":", (bookings_hotels_df.room_class.str.count(i).sum() / total_successful_bookings) * 100)
```

##### Mostly Elite Rooms are being booked, followed by Standard and Premium.

```python
#Average Daily Rate

average_daily_rate = total_revenue_realized / total_bookings
average_daily_rate
```

##### Average Revenue from a single booking is almost 12696 Rs.

```python
#Realization %

realization_per = 100 - (cancellation_per + no_show_bookings_per)
realization_per
```

##### Realization Percentage is Good but need to look out on Cancellation Percentage. Realization Percentage is the actual number of customer is a part of revenue.

```python
#Revenue per available room

revpar = total_revenue_realized / total_capacity
revpar
```

##### Average Revenue for a room available is almost 7347, which is Good.

```python
#Daily Booked Room Nights
#given number of days = 92

dbrn = total_bookings / 92
dbrn
```

##### Almost 1463 rooms are booked every night. Total rooms available every night are 232576, almost 231113 rooms are empty every night.
##### More customers are need to attracted by creative marketing techniques.

```python
#Daily Utilized Room Nights

durn = total_checked_out / 92
durn
```

##### Only 1026 rooms are actually used every night which is a very poor insight.

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

labels = ['revenue_generated', 'revenue_realized']
counts = [bookings_hotels_df['revenue_generated'].sum(), bookings_hotels_df['revenue_realized'].sum()]

sns.barplot(x = labels, y = counts, palette = 'rocket')

bookings_hotels_df['revenue_generated'].sum(), bookings_hotels_df['revenue_realized'].sum()
```

##### Total revenue realized is way lower than revenue generated. 29,87,74,986 is the difference between Revenue Generated and Revenue Realized.

```python
fig, axes = plt.subplots(4, 2, figsize = (18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')
sns.countplot(ax = axes[0, 0], x = 'property_name', data = df, palette = 'rocket', order = bookings_hotels_df['property_name'].value_counts().index);
sns.countplot(ax = axes[0, 1], x = 'booking_platform', data = df, palette = 'rocket', order = bookings_hotels_df['booking_platform'].value_counts().index);
sns.countplot(ax = axes[1, 0], x = 'category', data = df, palette = 'rocket', order = bookings_hotels_df['category'].value_counts().index);
sns.countplot(ax = axes[1, 1], x = 'city', data = df, palette = 'rocket', order = bookings_hotels_df['city'].value_counts().index);
sns.countplot(ax = axes[2, 0], x = 'room_class', data = df, palette = 'rocket', order = bookings_hotels_df['room_class'].value_counts().index);
sns.countplot(ax = axes[2, 1], x = 'booking_status', data = df, palette = 'rocket', order = bookings_hotels_df['booking_status'].value_counts().index);
sns.countplot(ax = axes[3, 0], x = 'mmm yy', data = df, palette = 'rocket', order = df['mmm yy'].value_counts().index);
sns.countplot(ax = axes[3, 1], x = 'week no', data = df, palette = 'rocket', order = df['week no'].value_counts().index);
axes[0][0].tick_params(labelrotation=45);
axes[0][1].tick_params(labelrotation=45);
```

##### 1. Atliq Palace, Atliq Exotica and Atliq City have most of the bookings.
##### 2. Others, MakeYourTrip and LogTrip attract most of the customers.
##### 3. Customers are most likely to book luxury hotel rather than business.
##### 4. Most the of bookings are from Mumbai, followed by Hyderabad, Banglore and Delhi.
##### 5. Customers are more likely to book an Elite room, followed by Standard, Premium and Presidential.
